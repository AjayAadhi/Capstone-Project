## Predicting the GDP of a Country and the contribution of factors

### Attributes in Dataset

#### Country => List of countries in the dataset
#### Region => The region in which the country is present
#### Population => Population of each country
#### Area => Total Area of the country in sq.mi.
#### Population Density => The density of population in each region
#### Coastline => The ratio of coast/area
#### Net Migration => The number of people migrating
#### Infant Mortality(per Thousand Births) => Infant mortality is the death of an infant before his or her first birthday.
#### The infant mortality rate is the number of infant deaths for every 1,000 live births.
#### GDP($ PER CAPTA) => Gdp of a country
#### Literacy => % of people who has the ability to read and write.
#### Phones(per 1000) => No of people who have Mobile phone per 1000 people
#### Arable(%) => (in farming) connected with growing crops for sale, not keeping animals.
#### Crops(%) => percentage of crops produced in the country
#### others(%) => represents the small scale industries
#### climate => shows the climatic condition in the region
#### Birthrate => The birth rate is the ratio between the number of live-born births in the year and the average total population of that year.
#### Deathrate => The Death rate is the ratio between the number of deaths in the year and the average total population of that year.
#### Agriculture =>  represents the agricultural contribution to gdp of a country
#### Industry  => represents the industrial contribution to gdp of a country
#### Service => reprresents the contribution provided by the servies done by the country

### Treating the missing values 

```python
for col in df.columns.values:
        if df[col].isnull().sum() == 0:
            continue
        if col == 'Climate':
             guess_values = df.groupby('Region')['Climate'].apply(lambda x: x.mode().max())
        else:
             guess_values = df.groupby('Region')[col].median()
        for region in df['Region'].unique():
             df[col].loc[(df[col].isnull())&(df['Region']==region)] = guess_values[region]
```

### Train and Test split

```python
train, test = train_test_split(df, test_size=0.3, random_state=0)#train test split
```

```python
training_features = ['Population', 'Area (sq. mi.)',
       'Pop. Density (per sq. mi.)', 'Coastline (coast/area ratio)',
       'Net migration', 'Infant mortality (per 1000 births)',
       'Literacy (%)', 'Arable (%)','Phones (per 1000)',
        'Crops (%)', 'Other (%)', 'Birthrate',
       'Deathrate', 'Agriculture', 'Industry', 'Service', 'Region_label',
       'Climate_label']
target = 'GDP ($ per capita)'
X_train = train[training_features]
Y_train = train[target]
X_test = test[training_features]
Y_test = test[target]
```

## LinearRegression

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
train_pred_Y = model.predict(X_train)
test_pred_Y = model.predict(X_test)
train_pred_Y = pd.Series(train_pred_Y.clip(0, train_pred_Y.max()), index=Y_train.index)
test_pred_Y = pd.Series(test_pred_Y.clip(0, test_pred_Y.max()), index=Y_test.index)

rmse_train = np.sqrt(mean_squared_error(train_pred_Y, Y_train))
msle_train = mean_squared_log_error(train_pred_Y, Y_train)
rmse_test = np.sqrt(mean_squared_error(test_pred_Y, Y_test))
msle_test = mean_squared_log_error(test_pred_Y, Y_test)
```

```python
print('rmse_train:',rmse_train,'msle_train:',msle_train)
print('rmse_test:',rmse_test,'msle_test:',msle_test)
```

```python
model.score(X_train,Y_train)
```
